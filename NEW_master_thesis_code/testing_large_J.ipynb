When increasing the synaptic strength J by a factor of 10, we should see more irregular spiking and an increase in the CV values.

We look at the CV distributions of the brunel model with rounded spike times, delays drawn from [1.0, 2.0], and resolution 1/8. First look at the CV values with our standard J value.

```python
import pandas as pd
#import nest
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import elephant
import quantities as pq
import neo
import sys
from scipy.stats import ks_2samp
plt.rcParams['figure.figsize'] = (14, 10)
import scipy.stats as stats
import itertools
import random
# If seaborn and elephant not available, run the sommands:
# !{sys.executable} -m pip install elephant
# !{sys.executable} -m pip install seaborn
```

```python
cv_exc_01 = pd.read_csv('C:\\Users\\Nida\\Documents\\NMBU\\master\\spike_data\\CV_distributions\\resolution_1_8\\brunel_round_1_2_exc_cv.csv',index_col=0)
```

```python
path = 'C:\\Users\\Nida\\Documents\\NMBU\\master\\spike_data\\brunel_10s\\resolution_1_8\\brunel_rounding_1_2\\testing_J_increase\\brunel_rounding_True_delay_1.0_2.0_seed_1_spikes_exc-12502-0.dat'

exc_spikes = pd.read_csv(r'{}'.format(path),skiprows=2,sep='\t')

def get_cvs(spike_data):
    """
    Get the CV for each neuron recorded. 
    
    CV = standard_deviation(ISIs)/mean(ISIs)
    """
    cvs = []
     
    spike_data = spike_data.sort_values(by='time_ms')
    grouped = spike_data.groupby(spike_data['sender'])

    for name, group in grouped:
            """
            Each group is senders and times for one value of senders. That is, we iterate through all 
            neurons. And the times for each neuron is in sorted order. Therefore, the cvs
            returned must have the same order. So cvs contain cv of neuron 1, then neuron 2 .... then neuron N.
            """
            t = np.asarray(group['time_ms'])
            spiketrain = neo.core.SpikeTrain(t * pq.ms, t_start=0*pq.ms, t_stop=10000*pq.ms)
            isi = elephant.statistics.isi(spiketrain)
            cv = elephant.statistics.cv(isi)
            cvs.append(cv)
            
    return cvs

cv_exc_02 = get_cvs(exc_spikes)
```

```python
# take average cv of cv_exc_01
cv_arr = np.array(cv_exc_01)
avg_cv = np.mean(cv_arr,axis=0)
```

```python
# plot the two against each other with ecdf plots

sns.ecdfplot(avg_cv,label='original synaptic strengths')
sns.ecdfplot(cv_exc_02,label='stronger synaptic connections')
plt.legend();
```

As excpected we see a much higher vlaue of the CV distribution when we increased J by a factor of 10. We plot them individually here for more clarity.
