Illustration of extreme behaviour in FR when resolution is 1/4.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#plt.rcParams['figure.figsize'] = (14, 10)
```

```python
cm = 1/2.54  # centimeters in inches
plt.rcParams['figure.figsize'] = (8*cm,7*cm)
```

```python
fig_path = 'C:\\Users\\Nida\\Documents\\NMBU\\master\\Figures'
```

```python
df = pd.read_pickle('C:\\Users\\Nida\\Documents\\NMBU\\master\\microcircuit_spike_data\\microcircuit_77175_df.pkl')
```

```python
n = 10
colors = plt.cm.viridis(np.linspace(0, 1, n))
```

```python
for i in range(1,11):
    p =[float(x)*1000 for x in df.loc['droop','FR'][4][i].values]
    plt.plot(np.sort(p), np.linspace(0, 1, len(p), endpoint=False),color=colors[i-1], label=f'seed {i}')
#plt.legend()
plt.xlabel('FR (spikes/s)',fontsize=6)
plt.ylabel('Proportion',fontsize=6);
plt.xticks(fontsize=7);
plt.yticks(fontsize=7);
plt.savefig(r'{}\res_1_4_FR_microcircuit_ecdf.pdf'.format(fig_path),bbox_inches='tight')
```

```python
for i in range(1,11):
    p =[float(x)*1000 for x in df.loc['droop','FR'][8][i].values]
    plt.plot(np.sort(p), np.linspace(0, 1, len(p), endpoint=False),color=colors[i-1])
    
plt.xlabel('FR (spikes/s)',fontsize=6)
plt.ylabel('Proportion',fontsize=6);
plt.xticks(fontsize=7);
plt.yticks(fontsize=7);
plt.savefig(r'{}\res_1_8_FR_microcircuit_ecdf.pdf'.format(fig_path),bbox_inches='tight')
```
