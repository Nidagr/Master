# Analyze the CV and FR distributions of microcircuit model 

We want to plot and examine the data of the microcircuit model, node population L5I. First we retrieve the multiindex dataframe of CV and FR distributions for all variations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (14, 10)
```

```python
df = pd.read_pickle('C:\\Users\\Nida\\Documents\\NMBU\\master\\microcircuit_spike_data\\microcircuit_77175_df.pkl')
```

As the probability of each delay value in the microcircuit model is tha same for resolution 1/2, we do not have a droop and equal version of this. Also notice that the probability of each delay value for inhibitory connections is the same when resolution is 1/4. Now droop and equal version of this. Only for excitatory connections when resolution is 1/4 and higher resolutions. We start by examining for the resolutoion 1/4.

## CV distributions

Cv distributions for all resolutions droop model.

```python
for i in range(1,11):
    cv = df.loc['droop',2,i]['CV']
    sns.ecdfplot(cv,color='r',alpha=0.5)

for i in range(1,11):
    cv = df.loc['droop',4,i]['CV']
    sns.ecdfplot(cv,color='g',alpha=0.5) 
    
for i in range(1,11):
    cv = df.loc['droop',8,i]['CV']
    sns.ecdfplot(cv,color='b',alpha=0.5) 
    
for i in range(1,11):
    cv = df.loc['droop',16,i]['CV']
    sns.ecdfplot(cv,color='m',alpha=0.5) 
    
for i in range(1,11):
    cv = df.loc['droop',32,i]['CV']
    sns.ecdfplot(cv,color='c',alpha=0.5)
    
for i in range(1,11):
    cv = df.loc['droop',64,i]['CV']
    sns.ecdfplot(cv,color='k',alpha=0.5) 
    
for i in range(1,11):
    cv = df.loc['droop',128,i]['CV']
    sns.ecdfplot(cv,color='y',alpha=0.5) 
    
for i in range(1,11):
    cv = df.loc['droop',256,i]['CV']
    sns.ecdfplot(cv,color='gold',alpha=0.5) 
    
# continuous values    
for i in range(1,11):
    cv = df.loc['continuous',8,i]['CV']
    sns.ecdfplot(cv,color='aqua',alpha=0.5)
    
plt.title('CV distributions for droop model all resolutions, and the CV distributions of continuous model.');
```
