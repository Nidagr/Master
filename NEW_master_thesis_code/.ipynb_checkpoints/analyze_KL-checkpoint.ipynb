When examining the CV, FR and CC distributions we could not find a difference in the models. But maybe if we look at the KL divergences as well there could be. We will use the KL divergence for continuous variables for the CV distributions as they are continuous. For the FR distributions we will use the KL divergence for discrete variable as the FR distributions were cleary discrete in the plots. Not sure which version to use for the CC distributions, theye look kinda continuous on distance, but look discrete up close.

```python
#code for the continuous KL divergences.
def ecdf(x,tot_min,tot_max):
    x = np.sort(list(x)+[tot_min,tot_max])
    u, c = np.unique(x, return_counts=True)
    n = len(x)
    y = (np.cumsum(c) - 0.5)/n
    def interpolate_(x_):
        yinterp = np.interp(x_, u, y, left=tot_min, right=tot_max)
        return yinterp
    return interpolate_

def cumulative_kl(x,y,fraction=0.5): 
    dx = np.diff(np.sort(np.unique(x)))
    dy = np.diff(np.sort(np.unique(y)))
    ex = np.min(dx)
    ey = np.min(dy)
    e = np.min([ex,ey])*fraction
    n = len(x)
    # total max
    max_x = max(x)
    max_y = max(y)
    tot_max = max(max_x,max_y)
    # total min
    min_x = min(x)
    min_y = min(y)
    tot_min = min(min_x, min_y)
    # send tot_max and tot_min to interpolation limits
    P = ecdf(x,tot_min,tot_max)
    Q = ecdf(y,tot_min,tot_max)
    
    # machine epsilon
    machine_eps = np.finfo('float').eps
    summ = 0
    for i in range(0,len(x)):
        if(abs(Q(x[i])-Q(x[i]-e))>3*machine_eps):
            summ += np.log((P(x[i]) - P(x[i]-e))/(Q(x[i]) - Q(x[i]-e)))
    # 1/n*summ - 1 converges to KL divergence    
    KL = (1./n)*summ - 1
    #KL = (1./n)*np.sum(np.log((P(x) - P(x-e))/(Q(x) - Q(x-e)))) - 1
    return KL

# if we want the KL divergence between each pair of simulations from one model, use this
# can be useful to run on true model to have a reference point
def kl(cv_list_exc):
    divs_exc = []
    for i in range(1,11):
        for j in range(1,i):
            kl_exc = cumulative_kl(cv_list_exc[i],cv_list_exc[j],fraction=0.05)
            divs_exc.append(kl_exc)
    return divs_exc

# if we want to compare all distributions from two different models.
def kl_two(value_list_1,value_list_2):
    # send inn value list on the form df.loc['droop',8]['CV']
    t1 = []
    for i in range(1,11):
        l = value_list_1[i].values.tolist()
        t1.append(l)

    t2 = []
    for i in range(1,11):
        l = value_list_2[i].values.tolist()
        t2.append(l)
    comb = list(itertools.product(t1, t2))
    kls = []
    for i in range(0,len(comb)):
        kl = cumulative_kl(comb[i][0],comb[i][1],fraction=0.05)
        kls.append(kl)
    return kls
```

To get an idea of how the continuous KL divergences should look for samples from the same distribution we generate some normal distribution samples and run the KL test on them.

```python
# create 10 normal distributions with 10 000 numbers, mean 0 and standard deviation 1
x = np.random.normal(0, 1, size=(10,10000))

divs_cont = []
for i in range(0,10):
        for j in range(0,i):
            kl = cumulative_kl(x[i],x[j],fraction=0.05)
            divs_cont.append(kl)
```

# FR

Seems like the FR distributions have around 24-29 unique values. Varying a little for which resolution we have. But we should use one choice of number of bins across all plots/calculations.

According to article from Rice one should use 43 bins in our case, but since we have so few distinct values, better to choose number of bins as number of unique values? ALternating for all models, between 24 and 29, lets choose something in the middle, lets say 27 bins.

```python
def compute_probs(data, n=10): 
    h, e = np.histogram(data, n)
    p = h/len(data)
    return e, p

def support_intersection(p, q): 
    sup_int = (
        list(
            filter(
                lambda x: (x[0]!=0) & (x[1]!=0), zip(p, q)
            )
        )
    )
    return sup_int

def get_probs(list_of_tuples): 
    p = np.array([p[0] for p in list_of_tuples])
    q = np.array([p[1] for p in list_of_tuples])
    return p, q

def kl_divergence(p, q): 
    return np.sum(p*np.log(p/q))


def compute_kl_divergence(train_sample, test_sample, n_bins): 
    """
    Computes the KL Divergence using the support 
    intersection between two different samples
    """
    e, p = compute_probs(train_sample, n=n_bins)
    _, q = compute_probs(test_sample, n=e)

    list_of_tuples = support_intersection(p, q)
    p, q = get_probs(list_of_tuples)
    
    return kl_divergence(p, q)

def kl_divergences(fr_lists, n_bins=10):
    """
    Calculate the KL divergence for each combination of pairs of simulations from the same model.
    Possible to give number of bins inn this function.
    """
    divs = []
    for i in range(1,11):
        for j in range(1,i):
            # elements on form number * 1/ms, make it float
            x = [float(e) for e in fr_lists[i]]
            y = [float(e) for e in fr_lists[j]]
            kl = compute_kl_divergence(x,y, n_bins=n_bins)
            divs.append(kl)
    return divs

def kl_div_two_models(value_list_1, value_list_2,n_bins=10):
    """
    Calculate KL divergence for each possible pair of simulations from two models.
    """
    # send inn value list on the form df.loc['droop',8]['CV']
    t1 = []
    for i in range(1,11):
        l = value_list_1[i].values.tolist()
        t1.append(l)

    t2 = []
    for i in range(1,11):
        l = value_list_2[i].values.tolist()
        t2.append(l)
    comb = list(itertools.product(t1, t2))

    kl_divs = []
    for i in range(0,len(comb)):
        kl = compute_kl_divergence(comb[i][0],comb[i][1],n_bins=n_bins)
        kl_divs.append(kl)
    return kl_divs
```
